```python
import numpy as np
import matplotlib.pyplot as plt
eps = 2
x = np.linspace(-5,5,50)
def getHuber(x,eps):
    if np.abs(x)<eps:
        return np.power(x,2)/2
    return eps*(np.abs(x)-eps/2)
yHub = []
for i in x:
    yHub.append(getHuber(i,eps))
yMSE = np.power(x,2)/2
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
plt.plot(x,yHub,label='Huber loss')
plt.plot(x,yMSE,label='MSE loss')
plt.legend()
plt.title(r'Huber vs MSE, $\epsilon=2$')
plt.savefig('huberLossExample.png',bbox_inches='tight')
plt.show()
```

```python
def getEI(x,eps):
    if np.abs(x)<eps:
        return 0
    return np.abs(x)
yEI = []
for i in x:
    yEI.append(getEI(i,eps))
yMSE = np.power(x,2)/2

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
plt.plot(x,yEI,label='Epsilon insensitive loss')
plt.plot(x,yMSE,label='MSE loss')
plt.legend()
plt.title(r'Epsilon insensitive vs MSE, $\epsilon=2$')
plt.savefig('epsilonInsensitiveLossExample.png',bbox_inches='tight')
plt.show()
```

```python
def getSEI(x,eps):
    if np.abs(x)<eps:
        return 0
    return np.abs(x)
yEI = []
for i in x:
    yEI.append(getSEI(i,eps)**2/2)
yMSE = np.power(x,2)/2

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
plt.plot(x,yEI,label='Squared epsilon insensitive loss')
plt.plot(x,yMSE,label='MSE loss')
plt.legend()
plt.title(r'Squared epsilon insensitive vs MSE, $\epsilon=2$')
plt.savefig('sqEpsilonInsensitiveLossExample.png',bbox_inches='tight')
plt.show()
```

```python
trEp = np.linspace(0,10,25)
const = 2*trEp**0
plt.plot(trEp,const,label="constant")
plt.ylabel('Learning rate')
plt.xlabel('iterations')
plt.title('Adaptive learning rates')
plt.yticks([0,1,2],[0,1,2])
plt.xticks([])
plt.legend()
plt.savefig('learningRates1.png',bbox_inches='tight')
plt.show()
```

```python
adp = [2]
for i in range(1,25):
    if i in [7,13,20,24]:
        adp.append(adp[-1]/5)
        continue
    adp.append(adp[-1])
plt.plot(trEp,const,label="constant")
plt.plot(trEp,adp,label="adaptive")
plt.ylabel('Learning rate')
plt.xlabel('iterations')
plt.title('Adaptive learning rates')
plt.yticks([0,1,2],[0,1,2])
plt.xticks([])
plt.legend()
plt.savefig('learningRates2.png',bbox_inches='tight')
plt.show()
```

```python
yPow = [2]
for i in range(1,25):
    yPow.append(2/i**2)
plt.plot(trEp,const,label="constant")
plt.plot(trEp,adp,label="adaptive")
plt.plot(trEp,yPow,label="Power, degree = 2")
plt.ylabel('Learning rate')
plt.xlabel('iterations')
plt.title('Adaptive learning rates')
plt.yticks([0,1,2],[0,1,2])
plt.xticks([])
plt.legend()
plt.savefig('learningRates3.png',bbox_inches='tight')
plt.show()
```

```python
inv = [2]
for i in range(1,25):
    inv.append(1/(0.5*(i+1)))
plt.plot(trEp,const,label="constant")
plt.plot(trEp,adp,label="adaptive")
plt.plot(trEp,yPow,label="Power, degree = 2")
plt.plot(trEp,inv,label="Inverse")
plt.ylabel('Learning rate')
plt.xlabel('iterations')
plt.title('Adaptive learning rates')
plt.yticks([0,1,2],[0,1,2])
plt.xticks([])
plt.legend()
plt.savefig('learningRates4.png',bbox_inches='tight')
plt.show()
```

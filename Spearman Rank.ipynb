```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
x = np.linspace(-5,5,50)
y = np.tanh(x)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
plt.plot(x,y)
plt.title("Pearson's r: "+str(round(pearsonr(x,y)[0],2))+", spearman rank: "+str(round(spearmanr(x,y)[0],2)))
ax.set_xticks([-2,1,0,1,2],[-2,1,0,1,2])
plt.yticks([-1,0,1],[-1,0,1])
ax = fig.add_subplot(122)
x = np.linspace(0,4,50)
y2 = np.power(x,2)
plt.plot(x,y2)
plt.title("Pearson's r: "+str(round(pearsonr(x,y2)[0],2))+", spearman rank: "+str(round(spearmanr(x,y2)[0],2)))
ax.set_xticks([0,1,2],[0,1,2])
plt.savefig("pearsonVsSpearman.png",bbox_inches='tight')
plt.show()
```
